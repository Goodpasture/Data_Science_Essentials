# Survived
Survived is a categorical variable.  Despite being represented in storage with a bit/integer, the value itself represents a boolean.  If Survived = 1, then the data in the observation is actually saying Survived = true.  If Survived = 0, then the data in the observation is actually saying Survived = false.

```python
#Starting with the base check
df.Survived.value_counts()
```

```python
#Can show Max, Min, Std, and Mean simply because it is stored as a bit
df.Survived.describe()
```

```python
#Checking if there are any missing values
df[df.Survived.isnull()]
```

```python
#Showing a histogram distribution
df.Survived.hist()
```

The above histogram help demonstrate that Survived is categorical since there can be only 2 possible values.
